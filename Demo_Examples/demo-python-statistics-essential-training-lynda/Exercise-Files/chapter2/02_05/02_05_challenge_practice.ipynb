```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
tb = pd.read_csv('tb.csv')
tb.head()
```

```python
tb.columns
```

```python
tb_melt= tb.melt(['country', 'year'], ['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544', 'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534', 'f3544', 'f4554', 'f5564', 'f65', 'fu'], 'sexage', 'cases')
```

```python
tb_melt.head()
```

```python
tb_melt['sex'] = tb_melt['sexage'].apply(lambda s: s[0])
tb_melt['age'] = tb_melt['sexage'].apply(lambda s: s[1:])

def add_dash(s):
    l = len(s)
    if l == 1:
        return s
    elif l == 2:
        return s[0] + '-' + s[1] 
    elif l == 3:
        return s[0] + '-' + s[1:]
    elif l == 4:
        return s[:1] + '-' + s[2:]
    
tb_melt['age'] = tb_melt['age'].apply(lambda s: add_dash(s))
```

```python
final = tb_melt.dropna(subset=['cases'])
final.head()
```

```python
final = final.sort_values(['country', 'year', 'age', 'sex'])
final.head()
```

```python
final = final[['country', 'year', 'age', 'sex', 'cases']]
final = final.reset_index()
```

```python
final.head()
```

```python
final.drop(['index'], axis=1, inplace=True)
final.head()
```

```python
final.to_csv('final_training.csv', index=False)
```
