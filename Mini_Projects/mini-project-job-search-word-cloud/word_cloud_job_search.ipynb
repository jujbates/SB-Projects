```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load in the dataframe
df = pd.read_csv("datasets/job.csv")
```

```python
df.head()
```

```python
print("Observations: {}".format(df.shape[0]))
print("Features: {}".format(df.shape[1]))
```

```python
# Start with one review:
text = df.Requirements[0]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("img/job_first_review.png")
```

```python
# text_r = " ".join(review for review in df.Requirements)
df.Description = df.Description.fillna(' ')
text_d = " ".join(review for review in df.Description)
text = text_r + ' ' + text_d
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["looking", "e", "g", "key", "work", "flavors", "finish"
                  , 'will', 'make', 'including'
                  , 'use', 'help', 'including'
                  , 'maing', 'focused', ])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wc = WordCloud(background_color="white", max_words=50,
               stopwords=stopwords, contour_width=3).generate(text)


# Display the generated image:
# the matplotlib way:
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()

```
